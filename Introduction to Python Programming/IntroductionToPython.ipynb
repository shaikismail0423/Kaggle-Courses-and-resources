```python
def get_pay(num_hours):
    # Pre-tax pay, based on receiving $15/hour
    pay_pretax = num_hours * 15
    # After-tax pay, based on being in 12% tax bracket
    pay_aftertax = pay_pretax * (1 - .12)
    return pay_aftertax
```

If you multiply a string by a boolean with value True, it just returns that same string. And if you multiply a string by a boolean with value False, it returns an empty string (or a string with length zero).

## Conditions and Conditional Statements

For a list of common symbols you can use to construct conditions, check out the chart below:

- `==`   equals
- `!=`   does not equal
- `<`    less than
- `<=`   less than or equal to
- `>`    greater than
- `>=`   greater than or equal to


```python
def evaluate_temp(temp):
    # Set an initial message
    message = "Normal temperature."
    # Update value of message only if temperature greater than 38C
    if temp > 38:
        message = "Fever!"
    return message
```

```python
def evaluate_temp_with_else(temp):
    if temp > 38:
        message = "Fever!"
    else:
        message = "Normal temperature."
    return message
```

```python
evaluate_temp_with_else(12)
```

### Conditional Statements

```python
def evaluate_temp_with_elif(temp):
    if temp > 38:
        message = "Fever!"
    elif temp > 35:
        message = "Normal temperature."
    else:
        message = "Low temperature."
    return message
```

```python
evaluate_temp_with_elif(36)
```

```python
def get_dose(weight):
    # Dosage is 1.25 ml for anyone under 5.2 kg
    if weight < 5.2:
        dose = 1.25
    elif weight < 7.9:
        dose = 2.5
    elif weight < 10.4:
        dose = 3.75
    elif weight < 15.9:
        dose = 5
    elif weight < 21.2:
        dose = 7.5
    # Dosage is 10 ml for anyone 21.2 kg or over
    else:
        dose = 10
    return dose
```

```python
print(get_dose(12))
```

## Intro to Lists

When doing data science, you need a way to organize your data so you can work with it efficiently. Python has many data structures available for holding your data, such as lists, sets, dictionaries, and tuples.

Python uses zero-based indexing. For instance, a list will make it easier to:

- Hold items with any data type, including booleans, integers, and floats.
- Get an item at a specified position (first, second, third, etc.).
- Check the number of items.
- Add and remove items.

We can refer to any item in the list according to its position in the list (first, second, third, etc.). This is called indexing.

## Slicing

You can also pull a segment of a list, such as the first three entries or the last two entries. This is called slicing.

- To pull the first x entries, you use `[:x]`.
- To pull the last y entries, you use `[-y:]`.

## min() and max()

You can also get the minimum with `min()` and the maximum with `max()`.

## sum()

To add every item in the list, use `sum()`
